```python
import os
import numpy as np
import pandas as pd
```

```python
training_path = '/content/drive/MyDrive/Newcastle University/Final Project and Dissertation/Dataset/std_shape/1.training'
```

```python
images = os.listdir(training_path)
print(len(images))
images[:10]
```

Removing images with more than one class label

```python
a = '[1, 1, 0]'
b = '[1, 0, 1]'
c = '[0, 1, 1]'
d = '[1, 1, 1]'
```

```python
# images with more than one class label
images_2 = [x for x in images if ((a in x) or (b in x) or (c in x) or (d in x))]
len(images_2)
```

```python
# images with only one class label, images not in images_2 above
images_3 = [x for x in images if x not in images_2]
print(len(images_3))
```

```python
# the final vesion to get images with only one class label from original list
images_4 = [x for x in images if ((a not in x) and (b not in x) and (c not in x) and (d not in x))]
len(images_4)
```

```python
image_paths = [training_path + '/' + x for x in images_4]
print(len(image_paths))
image_paths [1:5]
```

```python
# extracting labels from the image paths
labels = [x[-13:-4].strip('][').split(', ') for x in image_paths]
print(len(labels))
```

```python
labels = [list(map(int, x)) for x in labels]
len(labels)
```

# Creating the dataframe

The below dataframe is for training images belong to only one class. The data_frame for images belonginh to more than one class will be considered once the results from the current skew of the data is obtained

```python
images_dataframe = pd.DataFrame(list(zip(image_paths, labels)),
                                columns = ['image_path', 'labels'])

images_dataframe.head()
```
