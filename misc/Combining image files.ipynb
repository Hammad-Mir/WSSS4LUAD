```python
import os
import cv2
from matplotlib import pyplot as plt
```

```python
data_path = '/content/drive/MyDrive/Newcastle University/Final Project and Dissertation/Dataset/Original'
dest_path = '/content/drive/MyDrive/Newcastle University/Final Project and Dissertation/Dataset/std_shape/data_border_constant'
splits = os.listdir(data_path)
```

```python
# Works only for training set due to image dimensions
for sp in splits:
  data_path_1 = data_path + '/' + sp + '/' + sp
  print(data_path_1)
  image_list = os.listdir(data_path_1)
  image_list = [x for x in image_list if x.endswith('.png')]
  for im in image_list:
    im_path = data_path_1 + '/' + im
    #print(im_path)
    img = cv2.imread(im_path)
    #plt.imshow(im)
    image = cv2.copyMakeBorder(img, (300-img.shape[0])//2, (300-img.shape[0])//2, (300-img.shape[1])//2, (300-img.shape[1])//2, cv2.BORDER_CONSTANT, value = 255)
    image = cv2.resize(image, (224, 224))
    #plt.imshow(image)
    d_path = dest_path + '/' + str(sp) + '/' + str(im)
    #print(d_path)
    cv2.imwrite(d_path, image)
```

```python
def check_and_pad(im_src, im_dest):
  image_list = os.listdir(im_src)
  for im in image_list:
    im_path = im_src + '/' + im
    print(im_path)
    image = cv2.imread(im_path)

    if image.shape[0]>500 & image.shape[1] > 500:
      d_path = im_dest + '/' + im
      #print(d_path)
      cv2.imwrite(d_path, image)
      
    else:
      image = cv2.copyMakeBorder(image, (500-image.shape[0])//2, (500-image.shape[0])//2, (500-image.shape[1])//2, (500-image.shape[1])//2, cv2.BORDER_REFLECT101)
      #image = cv2.resize(image, (224, 224))
      d_path = im_dest + '/' + im
      #print(d_path)
      cv2.imwrite(d_path, image)
```
