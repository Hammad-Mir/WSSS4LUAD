```python
import os
import cv2
import time
from matplotlib import pyplot as plt
```

```python
def check_and_pad(im_src, im_dest):
  image_list = os.listdir(im_src)
  for im in image_list:
    im_path = im_src + '/' + im
    print(im_path)
    image = cv2.imread(im_path)
    #plt.imshow(image)
    #time.sleep(0.1)

    if (image.shape[0] > 500) and (image.shape[1] > 500):
      d_path = im_dest + '/' + im
      #print(d_path)
      cv2.imwrite(d_path, image)
      continue
      
    else:
      image = cv2.copyMakeBorder(image, (500-image.shape[0])//2, (500-image.shape[0])//2, (500-image.shape[1])//2, (500-image.shape[1])//2, cv2.BORDER_REFLECT101)
      image = cv2.resize(image, (500, 500))
      d_path = im_dest + '/' + im
      #print(d_path)
      cv2.imwrite(d_path, image)
```
