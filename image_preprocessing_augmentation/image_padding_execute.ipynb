```python
import os
import cv2
import argparse
import numpy as np
import skimage as si
from matplotlib import pyplot as plt
```

```python
data_path = '/home/zephyr/Desktop/Newcastle_University/11_FP_D/Dataset/Dataset/1.training/1.training/'
dest_path = '/home/zephyr/Desktop/Newcastle_University/11_FP_D/Dataset/data_border_reflect/'
dest_path_2 = '/home/zephyr/Desktop/Newcastle_University/11_FP_D/Dataset/data_border_constant/'
```

```python
os.listdir(data_path)
```

```python
image_list = os.listdir(data_path)
for im in image_list:
  im_path = data_path + '/' + im
  #print(im_path)
  img = cv2.imread(im_path)
  #plt.imshow(im)
  image = cv2.copyMakeBorder(img, (300-img.shape[0])//2, (300-img.shape[0])//2, (300-img.shape[1])//2, (300-img.shape[1])//2, cv2.BORDER_REFLECT101)
  image = cv2.resize(image, (224, 224))
  #plt.imshow(image)
  d_path = dest_path + '/' + str(im)
  #print(d_path)
  cv2.imwrite(d_path, image)
```
