```python
import os
import cv2
import torch
import numpy as np
import torchsummary

from matplotlib import pyplot as plt
import segmentation_models_pytorch as smp
from torchvision import transforms, utils

%matplotlib inline
```

```python
device = 'cuda:0' if torch.cuda.is_available() else 'cpu'
device
```

```python
train_image_path = '/home/zephyr/Desktop/Newcastle_University/11_FP_D/Dataset/Dataset/1.training/1.training/436219-7159-48057-[1, 0, 0].png'
valid_img_path = '/home/zephyr/Desktop/Newcastle_University/11_FP_D/Dataset/WSSS4LUAD/2.validation/2.validation/img/02.png'
```

```python
train_image = cv2.resize(cv2.imread(train_image_path), (224, 224))
valid_image = cv2.resize(cv2.imread(valid_img_path), (224, 224))
```

```python
fig = plt.figure(figsize=(10, 10))

ax = fig.add_subplot(1, 2, 1)
plt.imshow(train_image)

ax = fig.add_subplot(1, 2, 2)
plt.imshow(valid_image)
```

```python
def preprocess_image(img: np.ndarray, mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225], resized = (224, 224)) -> torch.Tensor:
  
  preprocessing = transforms.Compose([
                                      transforms.ToTensor(),
                                      transforms.Resize((224, 224)),
                                      transforms.Normalize(mean, std),
                                      ])
  return preprocessing(img.copy()).unsqueeze(0)
```

```python
train_image_tensor = preprocess_image(train_image)
valid_image_tensor = preprocess_image(valid_image)
```

```python
model_weights = ['/home/zephyr/Desktop/Newcastle_University/11_FP_D/Models/' +
                 x for x in os.listdir('/home/zephyr/Desktop/Newcastle_University/11_FP_D/Models')
                 if 'deeplab' in x and '50' in x]
model_weights.sort()
model_weights
```

```python
model = smp.DeepLabV3Plus(encoder_name='resnet50', classes=4, activation=None, encoder_weights=None, ).to(device)
```

```python
model.load_state_dict(torch.load(f=model_weights[1], map_location=device))
```

```python
%%capture
model.eval()
```

```python
train_out = model(train_image_tensor.to(device))
valid_out = model(valid_image_tensor.to(device))
```

```python
t_0 = torch.nn.functional.softmax(train_out, dim=1)
t_0 = np.transpose(t_0.squeeze(0).cpu().detach().numpy(), (1, 2, 0))

t_1 = torch.nn.functional.softmax(valid_out, dim=1)
t_1 = np.transpose(t_1.squeeze(0).cpu().detach().numpy(), (1, 2, 0))
```

```python
fig = plt.figure(figsize=(40, 20))

ax = fig.add_subplot(1, 4, 1)
plt.imshow(t_0[:, :, 0])

ax = fig.add_subplot(1, 4, 2)
plt.imshow(t_0[:, :, 1])

ax = fig.add_subplot(1, 4, 3)
plt.imshow(t_0[:, :, 2])

ax = fig.add_subplot(1, 4, 4)
plt.imshow(t_0[:, :, 3])
```

```python
fig = plt.figure(figsize=(40, 20))

ax = fig.add_subplot(1, 4, 1)
plt.imshow(t_1[:, :, 0])

ax = fig.add_subplot(1, 4, 2)
plt.imshow(t_1[:, :, 1])

ax = fig.add_subplot(1, 4, 3)
plt.imshow(t_1[:, :, 2])

ax = fig.add_subplot(1, 4, 4)
plt.imshow(t_1[:, :, 3])
```
